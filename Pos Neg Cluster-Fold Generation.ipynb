```python
import csv
import numpy as np
import pandas as pd
import itertools
```

```python
pos_file = "pos_cluster_detail.csv"
neg_file = "all_neg_cluster_details.csv"
```

```python
global cluster_number
global header

cluster_number = []
header_array = []

file = open(pos_file, "r")
for line in file:
        #print(line)
        cluster_number.append(int(line.split('-')[1]))
        header_array.append(line.split('-')[2])
        
```

```python
for i in header_array:
    print(i)
```

```python
print(len(cluster_number), len(header_array))
```

```python
fold0 = []
fold1 = []
fold2 = []
fold3 = []
fold4 = []
```

```python
global count1
global count2
count1 = 0
count2 = 0

for i in range(len(cluster_number)):
    
    if(cluster_number[count1] % 5 == 0):
        print("FOLD IS: " + str(cluster_number[count1] % 5))
        fold0.append(header_array[count2])
        print("Cluster Number: " + str(cluster_number[count1]) + " And Header is: " + header_array[count2])
        
    if(cluster_number[count1] % 5 == 1):
        print("FOLD IS: " + str(cluster_number[count1] % 5))
        fold1.append(header_array[count2])
        print("Cluster Number: " + str(cluster_number[count1]) + " And Header is: " + header_array[count2])
      
    if(cluster_number[count1] % 5 == 2):
        print("FOLD IS: " + str(cluster_number[count1] % 5))
        fold2.append(header_array[count2])
        print("Cluster Number: " + str(cluster_number[count1]) + " And Header is: " + header_array[count2])
    
        
    if(cluster_number[count1] % 5 == 3):
        print("FOLD IS: " + str(cluster_number[count1] % 5))
        fold3.append(header_array[count2])
        print("Cluster Number: " + str(cluster_number[count1]) + " And Header is: " + header_array[count2])
       
        
    if(cluster_number[count1] % 5 == 4):
        print("FOLD IS: " + str(cluster_number[count1] % 5))
        fold4.append(header_array[count2])
        print("Cluster Number: " + str(cluster_number[count1]) + " And Header is: " + header_array[count2])
    
    count1 = count1 + 1
    count2 = count2 + 1 
```

```python
print(len(fold0), len(fold1), len(fold2), len(fold3), len(fold4))
```

```python
df0 = pd.DataFrame(np.array(fold0))
df1 = pd.DataFrame(np.array(fold1))
df2 = pd.DataFrame(np.array(fold2))
df3 = pd.DataFrame(np.array(fold3))
df4 = pd.DataFrame(np.array(fold4))
```

```python
df0.to_csv("Fold 0.csv", index = None)
df1.to_csv("Fold 1.csv", index = None)
df2.to_csv("Fold 2.csv", index = None)
df3.to_csv("Fold 3.csv", index = None)
df4.to_csv("Fold 4.csv", index = None)
```
